```python
def transfor_2_np(list_in, x, y, z, h):
    if x == 0 and y ==0:
        output = np.zeros([z, h])
        for i in range(len(list_in)):
            for j in range(1):
                output[i,j] = list_in[i]
    elif x ==0 and not y == 0:
        output = np.zeros([ y, z, h])
        for i in range(len(list_in)):
            for j in range(len(list_in[i])):
                if type(list_in[i][j]).__name__ == 'list' or type(list_in[i][j]).__name__ == 'ndarray':
                    for k in range(len(list_in[i][j])):
                        output[i][j][k] = list_in[i][j][k]
                else:
                    output[i][j][0] = list_in[i][j]          
    else:
        output = np.zeros([x, y, z, h])
        for i in range(len(list_in)):#100
            for j in range(len(list_in[i])):#10
                for k in range(len(list_in[i][j])):#15
                    if type(list_in[i][j][k]).__name__ == 'list' or type(list_in[i][j][k]).__name__ == 'ndarray':
                        for l in range(len(list_in[i][j][k])):
                            try:
                                output[i][j][k][l] = list_in[i][j][k][l]
                            except:
                                print(i,j,k,l)
                    else:
                        print(type(list_in[i][j][k]).__name__)
                        output[i][j][k][0] = list_in[i][j][k] 
    return output
```

```python
from gensim.models.word2vec import Word2Vec
word2vec_greedy = Word2Vec.load('./data/greedy_context_node_w2v_128').wv
vocab_greedy = word2vec_greedy.vocab
word2vec_combine = Word2Vec.load('./data/combine_context_node_w2v_128').wv
vocab_combine = word2vec_combine.vocab
```

```python
def list_add(a,b):
    c = []
    for i in range(len(a)):
        c.append(a[i]+b[i])
    return c
```

```python
def list_divide(a,b):
    c = []
    for i in range(len(a)):
        c.append(a[i]/b)
    return c
```

```python
greedy_lable = []
combine_lable = []

greedy_path_embedding = []
combine_path_embedding = []

target = []

for index, row in c_train_data.iterrows():
    long_path_greedy_context = eval(row["long_path_greedy_context"])
    long_path_combine_context = eval(row["long_path_combine_context"])
    bug = row["bug"]
    target.append([bug])
    
    function_paths_greedy = []
    greedy_path_lable = []
    for path in long_path_greedy_context:
        lable = path[1]
        greedy_path_lable.append(lable)
        path_embedding = []
        for token in path[0][0]:
            if type(token).__name__ == 'list':
                sum_avg = [0] * 128
                for word in token:
                    sum_avg =list_add(word2vec_greedy.syn0[vocab_greedy[word].index],sum_avg)
                if not len(token)==0:
                    sum_avg = list_divide(sum_avg,len(token))
                    path_embedding.append(sum_avg)
            else:
                path_embedding.append(word2vec_greedy.syn0[vocab_greedy[token].index])
#         if len(path_embedding)<20:
#             for padding in range(len(path_embedding),20):
#                 path_embedding.append([0]*128)
        function_paths_greedy.append(path_embedding)
#     if len(function_paths_greedy)<15:
#             for padding in range(len(function_paths_greedy),15):
#                 function_paths_greedy.append([[0]*128]*20)
#                 greedy_path_lable.append(0)
    greedy_path_embedding.append(function_paths_greedy)
    greedy_lable.append(greedy_path_lable)    
    
    function_paths_combine = []
    combine_path_lable = []
    for path in long_path_combine_context:
        lable = path[1]
        combine_path_lable.append(lable)
        path_embedding = []
        for token in path[0][0]:
            if type(token).__name__ == 'list':
                sum_avg = [0] * 128
                for word in token:
                    sum_avg =list_add(word2vec_combine.syn0[vocab_combine[word].index],sum_avg)
                if not len(token)==0:
                    sum_avg = list_divide(sum_avg,len(token))
                    path_embedding.append(sum_avg)
            else:
                path_embedding.append(word2vec_combine.syn0[vocab_combine[token].index])
#         if len(path_embedding)<20:
#             for padding in range(len(path_embedding),20):
#                 path_embedding.append([0]*128)
        function_paths_combine.append(path_embedding)
#     if len(function_paths_combine)<15:
#             for padding in range(len(function_paths_combine),15):
#                 function_paths_combine.append([[0]*128]*20)
#                 combine_path_lable.append(0)
    combine_path_embedding.append(function_paths_combine)
    combine_lable.append(combine_path_lable)
    
    if index%999==0:
        print(index)
```

```python
npp =  transfor_2_np(combine_path_embedding,12491,15,20,128)
np.save("./data/local_im_test_input_combine.npy", npp)
buggy_states_list = transfor_2_np(combine_lable,0,12491,15,1)
np.save("./data/local_im_test_label_combine.npy", buggy_states_list)
target_list = np.mat(target)
np.save("./data/local_im_test_output_combine.npy", target_list)
```
