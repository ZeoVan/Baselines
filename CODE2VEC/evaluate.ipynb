```python
import numpy as np
def getting_data(file_path, print_text):
    print("Loading " + print_text + " Data...")
    data = np.load(file_path,allow_pickle = True)
    print("Done")
    return data
input_data = getting_data("./predict.npy", "Input")
```

```python
input_data.shape
```

```python
input_data[5]
```

```python
lable = []
with open("./data/mydata/mydata.test.c2v") as f:
    lines = f.readlines()
    for line in lines:
        print()
        if line[:4] =="good":
            lable.append(0)
        else:
            lable.append(1)
```

```python
len(lable)
```

```python
predict = np.argmax(input_data, axis=1)
predict_after_threshold = np.int64(predict<2)
```

```python
lable
```

```python
predict
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(predict_after_threshold, predict_after_threshold).flatten().tolist()
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(
    lable, predict_after_threshold
).flatten().tolist()

fpr_nn = fp/(fp+tp)
fnr_nn = fn/(fn+tn)
r = tp/(tp+fn)
p = tp/(tp+fp)
f = (2*p*r)/(p+r)
acc = (tp+tn)/(tp+fp+fn+tn)
```

```python
print("fpr_nn",fpr_nn)
print("fnr_nn",fnr_nn)
print("r",r)
print("p",p)
print("f",f)
print("acc",acc)
```

```python
lable = []
with open("./data/mydata/mydata.test.c2v") as f:
    lines = f.readlines()
    for line in lines:
        print()
        if line[:4] =="good":
            lable.append(0)
        else:
            lable.append(1)
len(lable)
```

```python
predict = np.argmax(input_data2, axis=1)
predict_after_threshold = np.int64(predict==1)
```

```python
tn, fp, fn, tp = metrics.confusion_matrix(
    lable, predict_after_threshold
).flatten().tolist()

fpr_nn = fp/(fp+tp)
fnr_nn = fn/(fn+tn)
r = tp/(tp+fn)
p = tp/(tp+fp)
f = (2*p*r)/(p+r)
acc = (tp+tn)/(tp+fp+fn+tn)
print("fpr_nn",fpr_nn)
print("fnr_nn",fnr_nn)
print("r",r)
print("p",p)
print("f",f)
print("acc",acc)
```
